```python
import pandas as pd
import numpy as np
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import association_rules, apriori
```

```python
data = pd.read_csv("Order3.csv")
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
transactions=data.groupby('TransactionNo')['Items'].apply(list).reset_index(name = "Transaction_Items")
transactions
```

```python
te = TransactionEncoder()
te_array = te.fit_transform(transactions['Transaction_Items'])

data_encoded = pd.DataFrame(te_array, columns=te.columns_)
```

```python
freq_items = apriori(data_encoded, min_support = 0.01, use_colnames=True)
freq_items
```

```python
rules = association_rules(freq_items,metric="lift",min_threshold=1)
rules
```
