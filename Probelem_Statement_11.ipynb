```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
points = np.array([
     [0.1, 0.6], [0.15, 0.71], [0.08, 0.9], [0.16, 0.85],
    [0.2, 0.3], [0.25, 0.5], [0.24, 0.1], [0.3, 0.2]
])
```

```python
#initialize centroids
m1 = np.array([0.1,0.6])
m2 = np.array([0.3,0.2])
```

```python
kmeans = KMeans(n_clusters=2, init=np.array([m1,m2]), n_init=1)
kmeans.fit(points)
```

```python
cluster_labels = kmeans.predict(points)
cluster_labels
```

```python
plt.scatter(points[cluster_labels == 0, 0], points[cluster_labels == 0, 1], c="blue", label="Cluster 1")
plt.scatter(points[cluster_labels == 1, 0], points[cluster_labels == 1, 1], c="red", label="Cluster 1")

plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], marker = "*", s=150, label="Centroids")

plt.scatter(m1[0],m1[1],c="green", s=150, label="Intial 1")
plt.scatter(m2[1],m2[0],c="yellow", s=150, label="Intial 2")

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("KMeans Clustering")
plt.legend()
plt.grid(True)
plt.show()
```

```python
point_p6 = np.array([[0.25,0.5]])
clus_p6 = kmeans.predict(point_p6)
print(f"P6 belongs to cluster {clus_p6[0]+1}")

population_m2 = np.sum(cluster_labels == 1)
print(f"Population around m2 is {population_m2}")

centroids = kmeans.cluster_centers_
print(f"Centroid 1: {centroids[0]}")
print(f"Centroid 2: {centroids[1]}")
```
